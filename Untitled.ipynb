```python
# Create a clean copy for analysis
weather_df = df.copy()
# Sort by date
weather_df = weather_df.sort_values('dt_iso').reset_index(drop=True)
```

```python
weather_df['dt_iso'] = pd.to_datetime(weather_df['dt_iso'], utc=True)
```

```python

def create_time_features(df, date_col='dt_iso'):
    # Ensure the column is datetime type
    if not pd.api.types.is_datetime64_any_dtype(df[date_col]):
        df[date_col] = pd.to_datetime(df[date_col])
    
    # Now extract features
    df['hour'] = df[date_col].dt.hour
    df['day_of_week'] = df[date_col].dt.dayofweek
    df['day_of_month'] = df[date_col].dt.day
    df['month'] = df[date_col].dt.month
    df['year'] = df[date_col].dt.year
    
    # Cyclical features
    df['hour_sin'] = np.sin(2 * np.pi * df['hour']/24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour']/24)
    df['month_sin'] = np.sin(2 * np.pi * df['month']/12)
    df['month_cos'] = np.cos(2 * np.pi * df['month']/12)
    
    return df

# Now call the function
weather_df = create_time_features(weather_df)
```

```python
# Convert Kelvin to Celsius for better interpretability
weather_df['temp_c'] = weather_df['temp'] - 273.15
weather_df['temp_min_c'] = weather_df['temp_min'] - 273.15
weather_df['temp_max_c'] = weather_df['temp_max'] - 273.15

# Create lag features for time series
for lag in [1, 2, 3, 4, 5, 6, 12, 24]:
    weather_df[f'temp_lag_{lag}'] = weather_df['temp_c'].shift(lag)
    weather_df[f'pressure_lag_{lag}'] = weather_df['pressure'].shift(lag)
    weather_df[f'humidity_lag_{lag}'] = weather_df['humidity'].shift(lag)

# Create rolling statistics
weather_df['temp_rolling_mean_6h'] = weather_df['temp_c'].rolling(window=6).mean()
weather_df['temp_rolling_std_6h'] = weather_df['temp_c'].rolling(window=6).std()
weather_df['pressure_rolling_mean_12h'] = weather_df['pressure'].rolling(window=12).mean()
# Weather category encoding (simplified)
weather_categories = {
    'clear': 0,
    'clouds': 1,
    'rain': 2
}
weather_df['weather_category'] = weather_df['weather_main'].map(weather_categories).fillna(0)
# Drop rows with NaN values from lag features
weather_df = weather_df.dropna()

print(f"Data shape after feature engineering: {weather_df.shape}")
print(f"Features created: {len(weather_df.columns)}")
```

```python
# Prepare features for forecasting
# We'll forecast temperature for next 1, 3, and 7 days (24, 72, 168 hours)
target_col = 'temp_c'
feature_cols = [
    'pressure', 'humidity', 'wind_speed', 'wind_deg', 'clouds_all',
]

# Create target variables for different forecast horizons
for horizon in [1, 3, 7]:  # days
    hours = horizon * 24
    weather_df[f'temp_next_{horizon}d'] = weather_df[target_col].shift(-hours)

# Remove rows with NaN in targets
df_clean = weather_df.dropna(subset=[f'temp_next_{h}d' for h in [1, 3, 7]]).copy()

print(f"Data shape for modeling: {df_clean.shape}")

# Split data into train and test sets (time-series aware split)
train_size = int(len(df_clean) * 0.8)
train_df = df_clean.iloc[:train_size].copy()
test_df = df_clean.iloc[train_size:].copy()

print(f"Train size: {len(train_df)}")
print(f"Test size: {len(test_df)}")
# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_df[feature_cols])
X_test_scaled = scaler.transform(test_df[feature_cols])
# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train_df[feature_cols])
X_test_scaled = scaler.transform(test_df[feature_cols])
```

## Model training with hyperparameter tuning

```python
# Define models and hyperparameters
models = {
    'Random Forest': {
        'model': RandomForestRegressor(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [10, 20, None],
            'min_samples_split': [2, 5]
        }
    },
    'Gradient Boosting': {
        'model': GradientBoostingRegressor(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [3, 5]
        }
    },
        'XGBoost': {
        'model': xgb.XGBRegressor(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [3, 5, 7]
        }
    }
}

# Store results
results = {}
```

```python
# Train models for each horizon
for horizon in [1, 3, 7]:
    print(f"\n{'='*30}")
    print(f"FORECASTING FOR {horizon} DAY(S) AHEAD")
    print(f"{'='*30}")
    
    y_train = train_df[f'temp_next_{horizon}d']
    y_test = test_df[f'temp_next_{horizon}d']
    
    horizon_results = {}
    
    for model_name, model_config in models.items():
        print(f"\nTraining {model_name}...")
                # Use time series cross-validation
        tscv = TimeSeriesSplit(n_splits=3)
        
        # Grid search with time series cross-validation
        grid_search = GridSearchCV(
            estimator=model_config['model'],
            param_grid=model_config['params'],
            cv=tscv,
            scoring='neg_mean_squared_error',
            n_jobs=-1,
            verbose=0
        )
        
        grid_search.fit(X_train_scaled, y_train)
         # Best model
        best_model = grid_search.best_estimator_
        
        # Predictions
        y_pred = best_model.predict(X_test_scaled)
        
        # Calculate metrics
        mae = mean_absolute_error(y_test, y_pred)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        r2 = r2_score(y_test, y_pred)
        
        horizon_results[model_name] = {
            'model': best_model,
            'params': grid_search.best_params_,
            'mae': mae,
            'rmse': rmse,
            'r2': r2,
            'predictions': y_pred
             }
        
        print(f"  Best params: {grid_search.best_params_}")
        print(f"  MAE: {mae:.2f}C, RMSE: {rmse:.2f}C, R: {r2:.3f}")
    
    results[horizon] = horizon_results
```

```python
# Create forecast visualization
fig, axes = plt.subplots(3, 1, figsize=(15, 12))
fig.suptitle('Temperature Forecasts vs Actual Values', fontsize=16, fontweight='bold')

best_models = {}
for idx, horizon in enumerate([1, 3, 7]):
    ax = axes[idx]
    
    # Get best model for this horizon (lowest RMSE)
    horizon_results = results[horizon]
    best_model_name = min(horizon_results.items(), 
                         key=lambda x: x[1]['rmse'])[0]
    best_model_info = horizon_results[best_model_name]
    
    best_models[horizon] = {
        'name': best_model_name,
        'model': best_model_info['model'],
        'rmse': best_model_info['rmse']
      }
    
    # Plot actual vs predicted
    y_test = test_df[f'temp_next_{horizon}d'].values
    y_pred = best_model_info['predictions']
    dates = test_df['dt_iso'].values[:len(y_test)]
    
    ax.plot(dates, y_test, label='Actual', linewidth=2, alpha=0.7)
    ax.plot(dates, y_pred, label=f'Predicted ({best_model_name})', 
            linewidth=2, alpha=0.7, linestyle='--')
    
    ax.fill_between(dates, y_test, y_pred, alpha=0.2, 
                   label='Forecast Error')

    ax.set_xlabel('Date')
    ax.set_ylabel(f'Temperature (C)')
    ax.set_title(f'{horizon}-Day Forecast (RMSE: {best_model_info["rmse"]:.2f}C)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('forecast_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
